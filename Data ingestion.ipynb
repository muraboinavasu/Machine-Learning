```python
from pyspark.sql.functions import (
    col, when, trim, lower, udf, lit
)
from pyspark.sql.types import IntegerType

print("Load from Parquet")

df = spark.read.parquet("wildchat_raw.parquet")
print(f" Loaded: {df.count()} rows")


print(" Data Cleaning")


df_clean = df \
    .fillna({"country": "Unknown", 
             "state": "Unknown", 
             "language": "Unknown"}) \
    .dropDuplicates(["conversation_hash"]) \
    .filter(col("turn") > 0) \
    .filter(col("language").isNotNull()) \
    .withColumn("language", trim(col("language"))) \
    .withColumn("country", trim(col("country")))

print(f" After cleaning: {df_clean.count()} rows")
```

```python
df_clean = spark.read.parquet("wildchat_raw.parquet")
print(f" Loaded: {df_clean.count()} rows")
```

```python
print("STEP 2: Fix All Columns")

from pyspark.sql.functions import col, when, isnull
from pyspark.sql.types import StringType

df_fixed = df_clean \
    .withColumn(
        "timestamp",
        col("timestamp").cast(StringType())
    ) \
    .withColumn(
        "timestamp",
        when(col("timestamp") == "nan", None)
        .otherwise(col("timestamp"))
    ) \
    .withColumn(
        "country",
        when(
            isnull(col("country")) |
            (col("country") == "nan") |
            (col("country") == ""),
            "Unknown"
        ).otherwise(col("country"))
    ) \
    .withColumn(
        "state",
        when(
            isnull(col("state")) |
            (col("state") == "nan") |
            (col("state") == ""),
            "Unknown"
        ).otherwise(col("state"))
    ) \
    .withColumn(
        "language",
        when(
            isnull(col("language")) |
            (col("language") == "nan") |
            (col("language") == ""),
            "Unknown"
        ).otherwise(col("language"))
    )

print("Fixed!")


print("STEP 3: Verify")

for c in df_fixed.columns:
    null_count = df_fixed.filter(isnull(col(c))).count()
    nan_count = df_fixed.filter(
        col(c).cast(StringType()) == "nan"
    ).count()
    total = null_count + nan_count

    if total > 0:
        print(f"{c:20s} | NULL:{null_count} | nan:{nan_count}")
    else:
        print(f"{c:20s} | Clean!")


print("STEP 4: Save")

df_fixed.write.mode("overwrite").parquet("wildchat_clean.parquet")

print(f" Saved! Rows: {df_fixed.count()}")

df_clean = df_fixed
```

```python


print("STEP 2: Full Missing Value Check")


for c in df_clean.columns:
    
    null_count = df_clean.filter(isnull(col(c))).count()
    
    nan_count = df_clean.filter(
        col(c).cast(StringType()) == "nan"
    ).count()
    
    empty_count = df_clean.filter(
        col(c).cast(StringType()) == ""
    ).count()
    
    total_missing = null_count + nan_count + empty_count
    
    if total_missing > 0:
        print(f"{c:20s} | NULL:{null_count} | nan:{nan_count} | empty:{empty_count}")
    else:
        print(f"{c:20s} |  Fully Clean")
```

```python

print("FULL ROW DUPLICATE CHECK")


total_rows = df_clean.count()
distinct_rows = df_clean.distinct().count()

duplicate_rows = total_rows - distinct_rows

print(f"Total Rows      : {total_rows}")
print(f"Distinct Rows   : {distinct_rows}")
print(f"Duplicate Rows  : {duplicate_rows}")
```
